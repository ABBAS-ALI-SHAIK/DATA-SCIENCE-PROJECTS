```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('life_expectancy.csv')
df.size
```

```python
df.columns
```

```python
df.loc[:,'Indicator Code']
```

```python
df = df[df['Indicator Code'] == "SP.DYN.LE00.IN"]
df.head(3)
```

```python
df = df.drop(df.loc[:, '1960':'1979'].columns, axis=1)
df.head(3)
```

```python
df.rename(columns={'Country Name': 'country'}, inplace=True)
df.head(3)
```

```python
df.drop(['Country Code','Indicator Name', 'Indicator Code'], axis=1, inplace=True)
df.head(3)
```

```python
df.set_index('country', inplace=True)
df
```

```python
df.loc['Zambia', '1990']
```

```python
df.isnull().sum().sum()
```

```python
df.dropna(how='all', axis=1, inplace=True)
df.dropna(how='all', axis=0, inplace=True)
df.dropna(how='any', axis=0, inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
df['Mean'] = df.mean(axis=1)
df.head(3) 
```

```python
countries = ['Japan', 'Kenya', 'Russian Federation']
years = list(map(str, range(1995, 2007)))

df.loc[countries, years].transpose().plot(
    kind='barh', 
    color=['lightcoral', 'gray', 'skyblue'], 
    figsize=(20, 14), 
    stacked=True # Значение накапливается
)

plt.xlabel('Life expectancy')
plt.ylabel('Year')
title = plt.title('Life expectancy in Japan, Kenya, and Russia')
```

```python
countries = ['Japan', 'Kenya', 'Russian Federation']
years = list(map(str, range(1995, 2007)))

df.loc[countries, years].transpose().plot(
    kind='barh', 
    color=['lightcoral', 'silver', 'skyblue'], 
    figsize=(22, 14), 
    stacked=True) # Значение накапливается

plt.xlabel('Life expectancy')
plt.ylabel('Year')
title = plt.title('Life expectancy in Japan, Kenya, and Russia')

for index, year in enumerate(years):
    value = 0
    for k, country in enumerate(countries): 
        
        label = int(df.loc[country,  year]) 
        value += label 
        plt.annotate(label, xy=(value-3, index), color='black')
```

```python
circle_diagram_data = df.head(10) 
circle_diagram_data
```

```python
circle_diagram_data = circle_diagram_data.groupby('country').mean() 
circle_diagram_data 
```

```python
axis = circle_diagram_data['Mean'].plot(
    kind='pie',   
    figsize=(18, 8),
)
```

```python
circle_diagram_data = circle_diagram_data.head(7)
circle_diagram_data.index.to_list()
```

```python
explode = [0, 0.1, 0, 0, 0.1, 0.2, 0] 

colors = ['#FF1159', '#129912',
          '#1299BB', '#AB12AC', 
          '#C29CAA', '#BCBB11', '#62988A',]

regions = circle_diagram_data.index.to_list()


circle_diagram_data['Mean'].plot(
    kind='pie',
    labels=regions, 
    colors=colors,
    figsize=(18, 8),
    autopct='%1.1f%%', 
    shadow=True,       
    explode=explode,   
    )


plt.ylabel('') 
plt.axis('equal') 
legend = plt.legend(loc='lower right') 
```
